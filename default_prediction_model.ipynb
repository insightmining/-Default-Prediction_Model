```python
from sklearn.model_selection import train_test_split
#print(data)

X_train = data.drop(['index'], axis=1)
y_train = data_label['credit']
X_test = test.drop(['index'], axis=1)
y_test = test_label['credit']

X_train_split, X_val, y_train_split, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)
X_train_split, X_val, y_train_split, y_val
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import AdaBoostClassifier, GradientBoostingClassifier, RandomForestClassifier
import lightgbm as lgb
import xgboost as xgb
from sklearn.metrics import f1_score

# 각 분류 모델 초기화
logistic_clf = LogisticRegression(random_state=42)
adaboost_clf = AdaBoostClassifier(n_estimators=50, random_state=42)
gradient_boost_clf = GradientBoostingClassifier(n_estimators=100, random_state=42)
random_forest_clf = RandomForestClassifier(n_estimators=100, random_state=42)
xgboost_clf = xgb.XGBClassifier(n_estimators=100, random_state=42)
lightgbm_clf = lgb.LGBMClassifier(n_estimators=100, random_state=42)

# 모델 리스트
models = [logistic_clf, adaboost_clf, gradient_boost_clf, random_forest_clf, xgboost_clf, lightgbm_clf]

# 각 모델을 훈련하고 예측 확률을 얻음
prob_predictions = []
for model in models:
    model.fit(X_train, y_train)
    probs = model.predict_proba(X_val)[:, 1]  # 클래스 1의 확률
    prob_predictions.append(probs)

# 예측 확률의 평균 계산
average_probs = np.mean(prob_predictions, axis=0)

# 임계값을 기준으로 0 또는 1로 변환
threshold = 0.5
y_pred_class = (average_probs > threshold).astype(int)


# F1 점수 계산
f1 = f1_score(y_val, y_pred_class)
print(f'F1 Score: {f1}')
```

```python
# final DataFrame 초기화
final = pd.DataFrame()

# 각 모델의 예측 결과를 final DataFrame에 추가
for i, model in enumerate(models):
    # 1차원 배열로 변환하여 DataFrame에 추가
    final[f'model_{i+1}'] = model.predict(X_test).flatten()
final_predict = final.mean(axis=1)
final_result = (final_predict > threshold).astype(int)
# final DataFrame의 각 열에 대한 평균을 계산하여 최종 예측값 생성
test_label['credit'] = final_result

# test_label DataFrame에 index 열 추가
test_label['index'] = test['index']

# 예측 결과를 CSV 파일로 저장
test_label.to_csv('/content/drive/MyDrive/Colab Notebooks/test_label_1.csv', index=False)
```
